```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 

from tqdm import tqdm
 
webdriver_path = "C:\Program Files (x86)\chromedriver.exe"
url = "https://www.imdb.com/search/title/?genres=Thriller&explore=title_type,genres&ref_=tt_ov_inf"
driver = webdriver.Chrome(webdriver_path)
driver.get(url)
# time.sleep(10)
```

```python
#todo

# get movie names, and url 

# story line and genrea
 
    
def movie_details(movies):
    url = []
    title = []
    
    for movie in movies: 
        m = movie.find_element(By.CLASS_NAME, 'lister-item-content')
        movie_url = movie.find_element(By.TAG_NAME,'a').get_attribute('href')
        movie_title = m.find_element(By.TAG_NAME,'a').text
        
        url.append(movie_url)
        title.append(movie_title)
    
    return url, title

def movie_stats(url):
    
    
    movie_genres = []
    
    driver.get(url)
    driver.execute_script("window.scrollTo(0, 0)")
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, 3500)")
    time.sleep(8)
    
    movie_storyline = (driver.find_element(By.CLASS_NAME, 'ipc-html-content-inner-div').text)
    genres = driver.find_elements("xpath","//li[@data-testid='storyline-genres']/div/ul/li")
    
    for i in genres:
        movie_genres.append(i.text)
    
    return movie_storyline, movie_genres
    
```

```python
import json

for i in tqdm(range(len(movie_urls))):
    
    if( i > 2235):
        try:
            story, genres = movie_stats(movie_urls[i])

            movie_data.append({
                'title': movie_titles[i],
                'url' : movie_urls[i],
                'description' : story,
                'genres': genres
            })

            with open('imdb_data2_500_data_new__fastr_records.json', 'w') as f:
                json.dump(movie_data, f ,indent=4)


        except:
            time.sleep(2)
            print("skipped")
```
